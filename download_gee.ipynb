```python
import ee
import geemap
Map = geemap.Map(center=(30, 120), zoom=4)
roi = ee.FeatureCollection("projects/assimov/assets/GB")
Map.addLayer(roi, {}, "roi")
Map
```

```python
elevation = ee.Image("NASA/NASADEM_HGT/001").select('elevation').clip(roi)
```

```python
fishnet = geemap.fishnet(roi, rows=4, cols=4, delta=1)
style = {'color': 'blue', 'fillColor': '00000000'}
Map.addLayer(fishnet.style(**style), {}, 'Fishnet')
```

```python
fishnetsize=fishnet.size().getInfo()
featlist = fishnet.toList(fishnet.size())
out_dir = '/content/drive/MyDrive/dem/'
print(fishnetsize)
```

```python
for i in range(0,fishnetsize):
    fea=ee.Feature(featlist.get(i)).bounds()
    img=elevation.clip(fea.geometry())
    tifname = out_dir+str(i)+".tif"
    print(tifname)
    geemap.download_ee_image(img,filename=tifname,scale=30,region=fea.geometry(),crs='EPSG:4326')
```

```python
region = ee.FeatureCollection("users/yamiletsharon250/wuhan").geometry()
s2Img = ee.ImageCollection('COPERNICUS/S2_SR')
def rmCloudByQA(image):
  qa = image.select('QA60')
  cloudBitMask = 1 << 10
  cirrusBitMask = 1 << 11 
  mask =qa.bitwiseAnd(cloudBitMask).eq(0)and(qa.bitwiseAnd(cirrusBitMask).eq(0))
  return image.updateMask(mask); 
S2 = s2Img.filterDate('2021-01-01', '2022-01-01').filterBounds(region).filter(ee.Filter.lt('CLOUDY_PIXEL_PERCENTAGE', 20)).map(rmCloudByQA).median().select(["B2", "B3", "B4", "B8","B11","B12"]).int32().clip(roi)
rgbVis = {
  'min': 0,
  'max': 3000,
  'bands': ['B4', 'B3', 'B2'],
}
Map.addLayer (S2, rgbVis, 'RGB')
geemap.download_ee_image(S2,filename="wuhan.tif",region=region,scale=10,crs='EPSG:4326')
```

```python
FishNet = geemap.fishnet(roi, rows=3, cols=3, delta=1)
style = {'color': 'red', 'fillColor': '00000000'}
Map.addLayer(FishNet.style(**style), {}, 'FishNet')
def clipimage(fc):
    net=ee.Feature(fc)
    img=S2.clip(net)
    return img
ImageCol = FishNet.map(clipimage)
out_dir = 'D:\data\wuhan/'
geemap.download_ee_image_collection(ee.ImageCollection(ImageCol),out_dir=out_dir,region=region,scale=10,crs='EPSG:4326')
```

```python
start_date = '2019-01-01'
end_date = '2019-03-31'

l8 = (
    ee.ImageCollection('LANDSAT/LC08/C01/T1_TOA')
    .filterBounds(roi)
    .filterDate(start_date, end_date)
)
def maskL8(image):
  qa = image.select('BQA')
  mask = qa.bitwiseAnd(1 << 4).eq(0)
  return image.updateMask(mask)

def addNDVI(image):
    ndvi = image.normalizedDifference(['B5', 'B4']).rename('NDVI')
    return ndvi

l8nocloud = l8.map(maskL8)
median = l8nocloud.median()
geemap.download_ee_image(addNDVI(median),filename="ndvi.tif",scale=30,region=roi.geometry(),crs='EPSG:4326')
```
